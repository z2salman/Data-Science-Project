```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
plt.rc('figure', figsize=(50, 50))
plt.style.use('ggplot')
np.set_printoptions(precision=4)

%matplotlib inline

df= pd.read_csv('data.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(12,7))
sns.boxplot(x="Make", y="Engine HP", data=df[['Make','Year','Engine HP']].loc[((df['Make'] == 'Chevrolet')|
               (df['Make'] == 'Ford')|
               (df['Make'] == 'Toyota')|
               (df['Make'] == 'Dodge')|
               (df['Make'] == 'Nissan')|
               (df['Make'] == 'Honda')|
               (df['Make'] == 'Mazda'))])
plt.grid()
plt.show()
```

```python
# Engine HP of German Big 3 within the past 4 years
plt.figure(figsize=(12,7))
sns.boxplot(x="Make", y="Engine HP", data=df[['Make','Year','Engine HP']].loc[((df['Make'] == 'BMW')|
               (df['Make'] == 'BMW')|
               (df['Make'] == 'Mercedes-Benz')|
               (df['Make'] == 'Audi')) & (df['Year'] > 2013)])
plt.grid()
plt.show()
```

```python
# Prices of German Big 3 within the past 4 years
plt.figure(figsize=(12,7))
sns.boxplot(x="Make", y="MSRP", data=df[['Make','Year','MSRP']].loc[((df['Make'] == 'BMW')|
               (df['Make'] == 'BMW')|
               (df['Make'] == 'Mercedes-Benz')|
               (df['Make'] == 'Audi')) & (df['Year'] > 2013)])
plt.grid()
plt.show()
```

```python
# Changes in Engine HP of German Big 3 within the past 4 years - HP is showing a decline across all brands
plt.figure(figsize=(12,7))
sns.boxplot(x="Year", y="Engine HP", data=df[['Make','Year','Engine HP']].loc[((df['Make'] == 'BMW')|
               (df['Make'] == 'BMW')|
               (df['Make'] == 'Mercedes-Benz')|
               (df['Make'] == 'Audi')) & (df['Year'] > 2013) ], hue='Make')
plt.grid()
plt.show()
```

```python
# Variation in Prices of German Big 3 within the past 4 years
plt.figure(figsize=(12,7))
sns.boxplot(x="Year", y="MSRP", data=df[['Make','Year','MSRP']].loc[((df['Make'] == 'BMW')|
               (df['Make'] == 'BMW')|
               (df['Make'] == 'Mercedes-Benz')|
               (df['Make'] == 'Audi')) & (df['Year'] > 2013) ], hue='Make')
plt.grid()
plt.show()
```

```python
# Popularity of German Big 3 - Popularity doesn't correlate with price and HP. Other factors such as design of the car should be investigated
plt.figure(figsize=(12,7))
sns.barplot(x="Make", y="Popularity", data=df[['Make','Year','Popularity']].loc[((df['Make'] == 'BMW')|
               (df['Make'] == 'BMW')|
               (df['Make'] == 'Mercedes-Benz')|
               (df['Make'] == 'Audi')) ])
plt.grid()
plt.show()
```
